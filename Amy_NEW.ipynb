```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
from pprint import pprint
from colorama import Fore
from colorama import Style
import datetime
import json
x = datetime.datetime.now()

# Automatically set Data Frames to 2 decimal places
pd.options.display.precision = 2

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# print(x.strftime("%d""/""%m""/""%Y"))

from api_keys import api_key
```

```python
days365=pd.Series(pd.date_range('2019', freq='D', periods=365)).dt.strftime("%Y-%m-%d")
days365
```

```python
DAYS= []
for days in days365:
    DAYS.append(str(days))
print(DAYS)
```

```python
asteroid_id=[]
name=[]
asteroid_date=[]
absolute_magnitude_h=[]
diameter_max_km = []
diameter_min_km =[]
diameter_min_km = []
is_potentially_hazardous_asteroid=[]
miss_distance_km = []
orbiting_body =[]
relative_velocity =[]
close_approach_date=[]
x=0



for date in DAYS: 
    url_1=(f"https://api.nasa.gov/neo/rest/v1/feed?start_date={date}&end_date={date}&api_key={api_key}")
    response=requests.get(url_1).json()
    
    
   

    inside=response["near_earth_objects"][date]

    try:
        for i in inside:
            inside2=inside[x]
            asteroid_date.append(date)
            asteroid_id.append(inside2['id'])


            name.append(inside2['name'])

            absolute_magnitude_h.append(inside2['absolute_magnitude_h'])

            diameter_max_km.append(inside2['estimated_diameter']['kilometers']['estimated_diameter_max'])

            diameter_min_km.append(inside2['estimated_diameter']['kilometers']['estimated_diameter_min'])
            is_potentially_hazardous_asteroid.append(inside2['is_potentially_hazardous_asteroid'])

            miss_distance_km.append(inside2['close_approach_data'][0]['miss_distance']['kilometers'])

            orbiting_body.append(inside2['close_approach_data'][0]['orbiting_body'])
            relative_velocity.append(inside2['close_approach_data'][0]['relative_velocity']['kilometers_per_second'])
            close_approach_date.append(inside2['close_approach_data'][0]["close_approach_date"])
            x=x+1
    except:
        print("End of data")
    x=0   
#     print("done...next day")
```
