```python
start = []
end = []

from datetime import date, timedelta

sdate = date(2019, 1, 1)   # start date
edate = date(2019, 12, 31)   # end date

delta = edate - sdate       # as timedelta

for i in range(int((delta.days /7)+1)):
    
    sday = sdate + timedelta(days=i*7)
    eday = sday + timedelta(days=6)
    
    if eday > edate:
        eday=edate
    start.append(sday)
    end.append(eday)
```
