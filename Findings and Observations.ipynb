# Findings and Observations



Whilst analysing data from NASA for Near Earth Objects and Asteroids we noticed that between 2000 and 2019 there has been a large, almost exponential, increase in the amount of objects observed since 2000.  From 101 in 2000 to 1142 in 2019! 

During this 19 year period, Average Distance Nominal was ~28% less in 2019,  and average diameter size decreased from just over 76% smaller. So it would seem that it isn't that there were more NEOs in the last 19 years, but that more smaller NEOs were observed and the objects came closer to Earth.

We chose to investigate 2019 further. Our team's questions were as follows:

### How many NEO's were recorded for the year of 2019?

The unique number of asteroids observed in 2019, that were Near Earth Objects were: 4510

### What is the biggest NEO recorded?

The largest recorded Near Earth Object was observed on 15th January 2019, its name was 433 Eros (A898 PA) (Asteroid ID :2000433). 
It's estimated maximum diameter was 4.94e+01 km's wide (49.4kms)

### Does size have an impact on speed of NEO's?
